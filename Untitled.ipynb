### Correlation and Multicollinearity
In the next phase of our analysis, we focus on understanding the relationships between the features in the dataset by examining correlation and multicollinearity. Correlation analysis helps us identify how strongly pairs of variables are related, providing insights into potential patterns that could influence customer churn. However, when variables are highly correlated with each other, a phenomenon known as multicollinearity, it can lead to issues in predictive modeling by inflating variance and making it difficult to assess the individual impact of each feature. To visualize these relationships and detect multicollinearity, we will use a heatmap of the correlation matrix, allowing us to easily spot highly correlated features and make informed decisions on feature selection or transformation.

```python
# Calculate the correlation matrix and round to two decimal places
corr_matrix = df.corr().round(2)

# Generate a mask for the upper triangle of the heatmap
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Set up the matplotlib figure with a custom size
plt.figure(figsize=(14, 12))

# Define a custom colormap
custom_cmap = sns.diverging_palette(240, 10, n=9, as_cmap=True)

# Draw the heatmap with the mask, custom colormap, and additional formatting
sns.heatmap(corr_matrix, mask=mask, cmap=custom_cmap, vmax=1, vmin=-1, center=0, annot=True,
            fmt=".2f", square=True, linewidths=1.5, cbar_kws={"shrink": 0.6, "aspect": 30}, 
            annot_kws={"size": 10, "weight": "bold"}, linecolor="black")

# Add a title with custom font size
plt.title('Correlation Matrix Heatmap', fontsize=16, fontweight='bold', color='darkblue')
```

### Model Perfomance 

To start, we will make a function, evaluate_model function that trains a classification model, calculates key performance metrics (precision, recall, accuracy, F1 score, and negative predictive value), and visualizes the results using a confusion matrix. It also performs 5-fold cross-validation to assess the model's stability and stores all metrics in a list for easy comparison. Finally, it prints a summary of the model's performance, providing a comprehensive evaluation of the trained model.

```python

# Custom function to evaluate and report model performance
def evaluate_model(model, X_train, y_train, results_list, dataset_label, model_label):
    """
    Trains a model, evaluates it on the training data, and stores the metrics.

    Args:
        model: An instance of an sklearn classifier.
        X_train: Features for training.
        y_train: Target variable for training.
        results_list: List to append the performance metrics.
        dataset_label: A label to identify the dataset used (e.g., 'Baseline', 'Resampled').
        model_label: A label to identify the model (e.g., 'LogReg', 'DecisionTree').

    Returns:
        The trained model instance.
    """
    # Fit the model
    model.fit(X_train, y_train)

    # Generate predictions
    y_pred_train = model.predict(X_train)

    # Confusion matrix
    conf_matrix = confusion_matrix(y_train, y_pred_train)
    tn, fp, fn, tp = conf_matrix.ravel()

    # Performance metrics
    precision = round(precision_score(y_train, y_pred_train), 2)
    recall = round(recall_score(y_train, y_pred_train), 2)
    accuracy = round(accuracy_score(y_train, y_pred_train), 2)
    f1 = round(f1_score(y_train, y_pred_train), 2)
    negative_pred = round(fn / (fn + fp), 2) if (fn + fp) != 0 else 0

    # Cross-validation for additional insights
    scoring = {'accuracy': 'accuracy', 'precision': 'precision', 'recall': 'recall'}
    cv_scores = cross_validate(model, X_train, y_train, scoring=scoring, cv=5, return_train_score=True)

    # Store results
    results_list.append({
        'Precision': precision,
        'Recall': recall,
        'Accuracy': accuracy,
        'F1 Score': f1,
        'Negative Predictive Value': negative_pred,
        'True Negative': tn,
        'False Positive': fp,
        'False Negative': fn,
        'True Positive': tp,
        'Dataset': dataset_label,
        'Model': model_label,
        'Stage': 'Train'
    })

    # Display results
    print(f"[{dataset_label} | {model_label}]")
    print(f"Train Set: Precision: {precision}, Recall: {recall}, Accuracy: {accuracy}, F1 Score: {f1}, NPV: {negative_pred}")
    print(f"Confusion Matrix:\n{conf_matrix}\n")
    print(f"Cross-Validation: Accuracy: {cv_scores['test_accuracy'].mean():.2f}, Precision: {cv_scores['test_precision'].mean():.2f}, Recall: {cv_scores['test_recall'].mean():.2f}\n")
    
    return model

# Initialize an empty list to collect model results
model_results = []
 
```

### Fine Tuning XGBoost

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Define the parameter grid
param_grid = {
    'max_depth': [4, 6, 8],
    'learning_rate': [0.1, 0.01, 0.001],
    'min_child_weight': [1, 5, 10],
    'booster': ['gbtree', 'gblinear', 'dart']
}

# Set up GridSearchCV
xgbc_new = GridSearchCV(XGBClassifier(random_state=42), param_grid, n_jobs=-1, scoring='recall')

# Define the evaluate_model function (basic version)
def evaluate_model(model, X_train, y_train, X_val, y_val, model_name):
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_val)
    
    # Calculate and print metrics
    print(f"Classification Report for {model_name}:")
    print(classification_report(y_val, y_pred))
    
    print(f"Confusion Matrix for {model_name}:")
    cm = confusion_matrix(y_val, y_pred)
    print(cm)
    
    # Plot the confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not Churned', 'Churned'])
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.show()

    return model

# Train new model using GridSearchCV
xgbc_new_gs = evaluate_model(xgbc_new, X_train_res, y_train_res, X_test, y_test, 'xgbc')

# Print the best parameters and best score
print("Best parameters found: ", xgbc_new_gs.best_params_)
print("Best recall score: ", xgbc_new_gs.best_score_)
```
