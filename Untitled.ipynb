### Correlation and Multicollinearity
In the next phase of our analysis, we focus on understanding the relationships between the features in the dataset by examining correlation and multicollinearity. Correlation analysis helps us identify how strongly pairs of variables are related, providing insights into potential patterns that could influence customer churn. However, when variables are highly correlated with each other, a phenomenon known as multicollinearity, it can lead to issues in predictive modeling by inflating variance and making it difficult to assess the individual impact of each feature. To visualize these relationships and detect multicollinearity, we will use a heatmap of the correlation matrix, allowing us to easily spot highly correlated features and make informed decisions on feature selection or transformation.

```python
# Calculate the correlation matrix and round to two decimal places
corr_matrix = df.corr().round(2)

# Generate a mask for the upper triangle of the heatmap
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Set up the matplotlib figure with a custom size
plt.figure(figsize=(14, 12))

# Define a custom colormap
custom_cmap = sns.diverging_palette(240, 10, n=9, as_cmap=True)

# Draw the heatmap with the mask, custom colormap, and additional formatting
sns.heatmap(corr_matrix, mask=mask, cmap=custom_cmap, vmax=1, vmin=-1, center=0, annot=True,
            fmt=".2f", square=True, linewidths=1.5, cbar_kws={"shrink": 0.6, "aspect": 30}, 
            annot_kws={"size": 10, "weight": "bold"}, linecolor="black")

# Add a title with custom font size
plt.title('Correlation Matrix Heatmap', fontsize=16, fontweight='bold', color='darkblue')
```

### Correlation with Churn:
The heatmap reveals that the churn variable shows a noticeable correlation with several features, particularly customer_service_calls (0.21), total_day_charge (0.21), and total_eve_charge (0.09). This indicates that customers who frequently contact customer service or incur higher charges for day and evening calls may be more likely to churn. These correlations could be significant predictors in our final model and warrant closer examination.

### Multicollinearity:
The heatmap also identifies perfect correlations (value of 1.00) between several pairs of features: total_day_minutes and total_day_charge, total_eve_minutes and total_eve_charge, total_night_minutes and total_night_charge, as well as total_intl_minutes and total_intl_charge. This is expected since charges are typically calculated based on the number of minutes spent on calls. However, this introduces multicollinearity into our model, which can distort the impact of individual variables and lead to overfitting. To mitigate this, we will drop the total_x_minutes columns (total_day_minutes, total_eve_minutes, total_night_minutes, and total_intl_minutes) from our dataset. Retaining only the corresponding total_x_charge columns should provide the same information without redundancy.

### Exploring the Relationship Between Features and Customer Churn 
we analyze the relationship between various features both categorical and numerical and customer churn. By examining how different attributes, such as geographic location, service plans, and usage patterns, are distributed among churned and non-churned customers, we aim to identify key factors that might influence a customer's decision to leave the company. This analysis includes visualizing the distribution of categorical variables as well as exploring correlations within numerical data. Through these insights, we can better understand the underlying patterns that contribute to customer churn, enabling the development of more effective retention strategies.

```python
categorical_features = ['state', 'international_plan', 'voice_mail_plan']

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Plotting the relationship between each categorical feature and Churn
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, hue='churn', data=df, palette='viridis')
    plt.title(f'Relationship between {feature} and churn')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.legend(title='churn')
    plt.xticks(rotation=45)  # Rotate x labels if needed
    plt.show()
```

It appears that there's a correlation between having an international plan and a higher likelihood of customer churn. This could suggest several possibilities:

- Dissatisfaction with International Plan: Customers who subscribe to the international plan might be facing issues or dissatisfaction with the service, leading them to churn.
- Pricing or Usage Issues: The cost of the international plan, usage limitations, or other factors could be contributing to customer churn.
- Alternative Providers: Customers with international needs might find more suitable or cost-effective options from other providers.

It appears that there's a correlation between having a voicemail plan and a lower likelihood of customer churn. This could suggest several possibilities:

- Value of Voicemail: Customers who subscribe to the voicemail plan might find it useful for various reasons, such as missed call notifications, message storage, or personal organization. This perceived value could contribute to higher customer satisfaction and retention.
- Bundling with Other Services: The voicemail plan might be bundled with other services, making it more attractive to customers and reducing the likelihood of them switching to a competitor.
- Customer Segmentation: The relationship between voicemail plan and churn could vary across different customer segments, such as demographics or usage patterns. Further analysis would be needed to identify specific segments that benefit most from this feature.

```python
numerical_features = [
    'account_length', 'number_vmail_messages', 'total_day_calls', 
    'total_eve_calls', 'total_night_calls', 'total_intl_calls', 
    'customer_service_calls'
]

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Loop through numerical features to create box plots and violin plots
for feature in numerical_features:
    plt.figure(figsize=(14, 6))
    
    # Create a subplot with 1 row and 2 columns for Box Plot and Violin Plot
    plt.subplot(1, 2, 1)
    sns.boxplot(x='churn', y=feature, data=df, palette='Set3')
    plt.title(f'Box Plot of {feature.replace("_", " ").title()} by churn')
    
    plt.subplot(1, 2, 2)
    sns.violinplot(x='churn', y=feature, data=df, palette='Set2')
    plt.title(f'Violin Plot of {feature.replace("_", " ").title()} by churn')
    
    plt.tight_layout()
    plt.show()
```

### One Hot Encoding

```python
# Defining the features (X) and target variable (y)
X = df.drop("churn", axis=1)
y = df['churn']
# Performing one-hot encoding on categorical features
categorical_cols = X.select_dtypes(exclude="number").columns
numerical_cols = X.select_dtypes(include="number").columns

X_categorical = X[categorical_cols]
X_numerical = X[numerical_cols]


# Initialize the OneHotEncoder with desired parameters
ohe = OneHotEncoder(sparse_output=False, handle_unknown='error', drop='first')

# Fit and transform the categorical data
X_categorical_ohe = pd.DataFrame(ohe.fit_transform(X_categorical), 
                         columns=ohe.get_feature_names_out(X_categorical.columns),
                         index=X_categorical.index)
X_categorical_ohe
```

```python
#merging the data set
X_merged = pd.concat([X_numerical, X_categorical_ohe], axis=1)
X_merged
```

```python
#transforms each feature to have a mean of 0 and a standard deviation of 1
#which is useful for many machine learning algorithms.
scaler = StandardScaler()

X_final = pd.DataFrame(scaler.fit_transform(X_merged), \
                               columns=(X_merged.columns) ,index=X_merged.index)

X_final
```
