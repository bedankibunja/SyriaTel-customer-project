### Correlation and Multicollinearity
In the next phase of our analysis, we focus on understanding the relationships between the features in the dataset by examining correlation and multicollinearity. Correlation analysis helps us identify how strongly pairs of variables are related, providing insights into potential patterns that could influence customer churn. However, when variables are highly correlated with each other, a phenomenon known as multicollinearity, it can lead to issues in predictive modeling by inflating variance and making it difficult to assess the individual impact of each feature. To visualize these relationships and detect multicollinearity, we will use a heatmap of the correlation matrix, allowing us to easily spot highly correlated features and make informed decisions on feature selection or transformation.

```python
# Calculate the correlation matrix and round to two decimal places
corr_matrix = df.corr().round(2)

# Generate a mask for the upper triangle of the heatmap
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Set up the matplotlib figure with a custom size
plt.figure(figsize=(14, 12))

# Define a custom colormap
custom_cmap = sns.diverging_palette(240, 10, n=9, as_cmap=True)

# Draw the heatmap with the mask, custom colormap, and additional formatting
sns.heatmap(corr_matrix, mask=mask, cmap=custom_cmap, vmax=1, vmin=-1, center=0, annot=True,
            fmt=".2f", square=True, linewidths=1.5, cbar_kws={"shrink": 0.6, "aspect": 30}, 
            annot_kws={"size": 10, "weight": "bold"}, linecolor="black")

# Add a title with custom font size
plt.title('Correlation Matrix Heatmap', fontsize=16, fontweight='bold', color='darkblue')
```

### Correlation with Churn:
The heatmap reveals that the churn variable shows a noticeable correlation with several features, particularly customer_service_calls (0.21), total_day_charge (0.21), and total_eve_charge (0.09). This indicates that customers who frequently contact customer service or incur higher charges for day and evening calls may be more likely to churn. These correlations could be significant predictors in our final model and warrant closer examination.

### Multicollinearity:
The heatmap also identifies perfect correlations (value of 1.00) between several pairs of features: total_day_minutes and total_day_charge, total_eve_minutes and total_eve_charge, total_night_minutes and total_night_charge, as well as total_intl_minutes and total_intl_charge. This is expected since charges are typically calculated based on the number of minutes spent on calls. However, this introduces multicollinearity into our model, which can distort the impact of individual variables and lead to overfitting. To mitigate this, we will drop the total_x_minutes columns (total_day_minutes, total_eve_minutes, total_night_minutes, and total_intl_minutes) from our dataset. Retaining only the corresponding total_x_charge columns should provide the same information without redundancy.
